As the target attribute is categorical, I used chi2 contingency method to determine if it is related to other attributes

* Dependent meaens they have a relation between the two attributes

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency 

stat, p, dof, expected =chi2_contingency(pd.crosstab(bank['term_deposit_subscribed'], bank['outcome_of_last_campaign']))
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

**Important note from the source of the dataset** : the _"last_contact_duration attribute"_ highly affects the output target (e.g., if duration=0 then y='no'). Yet, the duration is not known before a call is performed. Also, after the end of the call y is obviously known. Thus, this input should only be included for benchmark purposes and should be discarded if the intention is to have a realistic predictive model.

```python
stat, p, dof, expected =chi2_contingency(pd.crosstab(bank['term_deposit_subscribed'], bank['last_contact_duration']))
# interpret p-value
alpha = 0.05
print("p value is " + str(p))
if p <= alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```
